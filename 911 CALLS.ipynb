##### Defining new cell 'reason' from Title colum

```python
#df['title'].apply(lambda x : x.split(':')[0]).value_counts().head(10)

df['reason'] = df['title'].apply(lambda x : x.split(':')[0])
df.head()
```

```python
df['title'].apply(lambda x : x.split(':')[0]).value_counts().head(10)
```

```python
sns.countplot(x='reason',data=df)
plt.savefig('reasoncount.png',bbox_inches='tight')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] =  pd.to_datetime(df['timeStamp'])
```

##### Creating 3 columns for Hour, Month, and Day of the Week based on timeStamp

```python
df['Hour'] = df['timeStamp'].apply(lambda time : time.hour)
df['Month'] = df['timeStamp'].apply(lambda time:time.month)
df['DayOfWeek'] = df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df
```

##### Converting DayOfWeek to Sun-Sat from numbers

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['DayOfWeek'] = df['DayOfWeek'].map(dmap)
```

```python
df
```

```python
sns.countplot(x='DayOfWeek',data=df,hue='reason',palette='pastel',edgecolor='white')
plt.legend(loc="upper left",bbox_to_anchor=(1.0, 0.8))

plt.savefig('dayofWeek.png',bbox_inches='tight')
```

```python
sns.countplot(x='Month',data=df,hue='reason',palette='pastel',edgecolor='white')
plt.legend(loc=2,bbox_to_anchor=(1.0, 0.9))

plt.savefig('Month.png',bbox_inches='tight')
```

```python
byMonth = df.groupby('Month').count()
byMonth.head()
```

```python
#byMonth.dtypes

byMonth['twp'].plot()
plt.savefig('twpbyMonth.png',bbox_inches='tight')
```

```python
sns.lmplot(x='Month',y='desc',data=byMonth.reset_index())
plt.savefig('callspermonth.png')
```

```python
df['Date']= df['timeStamp'].apply(lambda t: t.date())
```

```python
plt.figure(figsize=(10,5))
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.savefig('twpByDate.png',bbox_inches='tight')
```

```python
plt.figure(figsize=(10,5))
df[df['reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
plt.savefig('EMSbydate.png',bbox_inches='tight')
```

```python
plt.figure(figsize=(10,5))
df[df['reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
plt.savefig('FirebyDate.png',bbox_inches='tight')
```

```python
plt.figure(figsize=(10,5))
df[df['reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
plt.savefig('trafficByDate.png',bbox_inches='tight')
```

##### Hour as column vs. DayOfWeek

```python
dayHour = df.groupby(by=['DayOfWeek','Hour']).count()['reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='viridis',linecolor='white',linewidths=1)
plt.savefig('newDF-HrByDayOfWeek.png',bbox_inches='tight')
```

```python
sns.clustermap(dayHour,cmap='viridis',linecolor='white',linewidths=1)
plt.savefig('clustermap.png',bbox_inches='tight')
```

##### Month as column vs. DayOfWeek

```python
dayMonth = df.groupby(by=['DayOfWeek','Month']).count()['reason'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='viridis',linecolor='white',linewidths=1)
plt.savefig('heatmap-MonthbyDayofweek.png',bbox_inches='tight')
```

```python
sns.clustermap(dayMonth,cmap='viridis',linecolor='white',linewidths=1)
plt.savefig('clustermap-MonthByDayofweek.png',bbox_inches='tight')
```
